# 12. Extracting Quote Containers

Each quote on Goodreads is stored within a dedicated HTML container.

BeautifulSoup is used to locate all quote blocks that contain author information, quote text, likes, and images.

```python
quotes= soup.find_all('div',class_="quoteDetails")
#quotes
```

```python
quotes_data1=[]
```

# 13. Extracting Required Information

For each quote block, the following attributes are extracted:

- Author Image URL
- Author Name
- Quote Text
- Number of Likes

The extracted data is stored in a Python list of dictionaries.

```python
for  quote in quotes:
    img=quote.find('img')
     #image = quote.find('img',class_='leftAlignedImage quoteAvatar')
    name =quote.find('span',class_='authorOrTitle').get_text(strip=True)
    quote_text = quote.find('div',class_='quoteText').get_text(strip=True)
    likes= quote.find('a',class_='smallText').get_text(strip=True)
    
    image =img['src'] if img and img.has_attr('src') else ""
 
    quotes_data1.append({
        'Image': image,
        'Name': name,
        'quote_text':quote_text,
        'likes':likes
    })
   
    
    
    
```
